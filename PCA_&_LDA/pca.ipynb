```python
import numpy as np
import matplotlib.pyplot as plt

# Create a synthetic dataset (2 features)
np.random.seed(0)
mean = [2, 3]
cov = [[3, 2], [2, 2]]  # covariance between features
X = np.random.multivariate_normal(mean, cov, 100)

# Plot original data
plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], alpha=0.7)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Original 2D Data")
plt.grid(True)
plt.show()
```

```python
# Subtract mean (centering)
X_meaned = X - np.mean(X, axis=0)
```

```python
cov_matrix = np.cov(X_meaned, rowvar=False)
print("Covariance matrix:\n", cov_matrix)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort eigenvectors by decreasing eigenvalues
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
# Take the first principal component (1D)
X_pca = X_meaned.dot(eigenvectors[:,0].reshape(-1,1))

# Plot PCA projection
plt.figure(figsize=(6,4))
plt.scatter(X_pca, np.zeros_like(X_pca), alpha=0.7)
plt.xlabel("Principal Component 1")
plt.title("Data projected onto 1st Principal Component")
plt.grid(True)
plt.show()
```
