```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("/Users/dharmajitbaro/Desktop/ailab/Student_Marks.csv")
```

```python
dataset.head
```

```python
x=dataset['time_study']
y=dataset['Marks']
```

```python
#initializing parameter
c=0 #intercept
m=0 #slope

learning_rate=0.01
epochs=1000
n=len(x)
```

```python
#training using gradient descent
for i in range(epochs):
    y_pred=m*x+c
    error=y-y_pred

    #gradients
    dm=(-2/n)*sum(x*error)          #derivative of loss fn wrt d & c
    dc=(-2/n)*sum(error)

    m=m-learning_rate*dm
    c=c-learning_rate*dc

    if i%100==0:
        mse=np.mean(error**2)
        print(f"Epoch {i}: m = {m:.4f}, c = {c:.4f}, MSE = {mse:.4f}")
```

```python
# Final values
print(f"Final: y = {m:.2f}x + {c:.2f}")
```

### Plotting Data Points and the Linear Regression Line

```python
plt.scatter(x, y, color='pink', label='Actual')
plt.plot(x, y_pred, color='red', label='Predicted Line')
plt.xlabel('Study_time')
plt.ylabel('Marks')
plt.legend()
plt.title('Linear Regression from Scratch')
plt.show()
```

### Calculating Loss Functions: RSS and RMS

```python
# RSS: Residual Sum of Squares
rss = np.sum((y - y_pred) ** 2)

# RMS: Root Mean Square Error
rms = np.sqrt(rss / n)

print(f"RSS = {rss:.4f}")
print(f"RMS = {rms:.4f}")
```

### Prediction for new data

```python
new_x = np.array([6, 7, 8])
new_y = m * new_x + c

print("Predictions for new inputs:")
for xi, yi in zip(new_x, new_y):
    print(f"x = {xi}, y = {yi}")
```
