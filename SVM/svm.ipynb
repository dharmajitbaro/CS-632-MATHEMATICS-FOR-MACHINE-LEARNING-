```python
#making simple binary classification with svm
from sklearn import svm
import numpy as np


X = np.array([
    [1, 2],
    [2, 3],
    [3, 3],
    [6, 5],
    [7, 8],
    [8, 8]
])

# Labels (0 or 1)
y = np.array([0, 0, 0, 1, 1, 1])

# Create SVM model
model = svm.SVC(kernel='linear')   # linear SVM

# Train the model
model.fit(X, y)

# Predict new points
test_points = np.array([
    [2, 2],
    [7, 7],
    [1,2]
])

predictions = model.predict(test_points)

print("Predictions for test points:", predictions)
```
