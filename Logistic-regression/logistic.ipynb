```python
from sklearn.linear_model import LogisticRegression
import numpy as np

# lets x is mark of student and y is pass[1] or fail[0]
X = np.array([[30], [40], [50], [60], [70], [80]])
y = np.array([0, 0, 0, 1, 1, 1])

# Create model
model = LogisticRegression()

# Train the model
model.fit(X, y)

# Predict for new marks
test_marks = np.array([[45], [75],[22],[99]])
predictions = model.predict(test_marks)

print("Predictions:", predictions)
```
