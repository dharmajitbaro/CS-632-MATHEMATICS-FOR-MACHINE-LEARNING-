## 1.Creating AND mp nuron

```python
def mpNuron(input,threashold):
    totalInput=sum(input)
    if totalInput >= threashold:
        return 1
    else:
        return 0

inputList=[
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

threashold=2

for input in inputList:
    print(f"{input[0]},{input[1]} ->{mpNuron(input,threashold)}")
```

## 2. Creating OR mp nuraon

```python
def orMpNuron(input,threashold):
    totalInput=sum(input)
    if totalInput>=threashold:
        return 1
    else:
        return 0
    
threashold=1

inputList=[
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

for input in inputList:
    print(f"{input[0]},{input[1]} ->{orMpNuron(input,threashold)}")
```

## 3. Creating NOR mp nuron

```python
def norMpNuron(inp):
    totalInput = sum(inp)
    
    # NOR: fires only when sum == 0
    if totalInput == 0:
        return 1
    else:
        return 0

inputList = [
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

for inp in inputList:
    print(f"{inp[0]},{inp[1]} -> {norMpNuron(inp)}")
```
